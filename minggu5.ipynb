```python
data_keuntungan = [
    {
        "tahun": 1995,
        "penjualan": 300,
        "operasional": 150,
        "pajak": 50,
        "pendapatan_lain": 135
    },
    {
        "tahun": 1997,
        "penjualan": 320,
        "operasional": 120,
        "pajak": 70,
        "pendapatan_lain": 105
    },
    {
        "tahun": 1991,
        "penjualan": 200,
        "operasional": 75,
        "pajak": 35,
        "pendapatan_lain": 85
    },
    {
        "tahun": 1992,
        "penjualan": 320,
        "operasional": 170,
        "pajak": 80,
        "pendapatan_lain": 235
    },
    {
        "tahun": 2001,
        "penjualan": 300,
        "operasional": 150,
        "pajak": 50,
        "pendapatan_lain": 135
    },
    {
        "tahun": 2003,
        "penjualan": 500,
        "operasional": 350,
        "pajak": 100,
        "pendapatan_lain": 55
    },
    {
        "tahun": 2005,
        "penjualan": 750,
        "operasional": 450,
        "pajak": 200,
        "pendapatan_lain": 155
    }
]
```

```python
data_grafik = {
    "tahun": [1991, 1992, 1995, 1997, 2001, 2003],
    "penjualan": [200, 320, 300, 320, 300, 500],
    "pajak": [35, 80, 50, 70, 50, 100],
    "operasional": [75, 170, 150, 120, 150, 350],
    "keuntungan": []
}
for data in data_keuntungan:
    if data["tahun"] in data_grafik["tahun"]:
        # Rumus keuntungan: penjualan - operasional - pajak + pendapatan_lain
        keuntungan = data["penjualan"] - data["operasional"] - data["pajak"] + data["pendapatan_lain"]
        data_grafik["keuntungan"].append(keuntungan)
```

```python
data_raw = {}
for dk in data_keuntungan:
    tahun = dk["tahun"]
    data_raw[tahun] = {
        "penjualan": dk["penjualan"],
        "operasional": dk["operasional"],
        "pajak": dk["pajak"],
        "pendapatan_lain": dk["pendapatan_lain"]
    }
data_raw
```

```python
data_tahun = list(data_raw.keys())
data_tahun.sort()
```

```python
data_penjualan = [data_raw[year]["penjualan"] for year in data_tahun]
data_pajak = [data_raw[year]["pajak"] for year in data_tahun]
data_operasional = [data_raw[year]["operasional"] for year in data_tahun]
data_pendapatan_lain = [data_raw[year]["pendapatan_lain"] for year in data_tahun]


data_grafik = {
    "tahun": data_tahun,
    "penjualan": data_penjualan,
    "pajak": data_pajak,
    "operasional": data_operasional,
    "pendapatan_lain": data_pendapatan_lain
}
data_grafik
```

```python
# Plot data penjualan
plt.plot(data_grafik["tahun"], data_grafik["penjualan"], label="Penjualan", marker='o')

# Plot data pajak
plt.plot(data_grafik["tahun"], data_grafik["pajak"], label="Pajak", linestyle='--', marker='x')

# Plot data operasional
plt.plot(data_grafik["tahun"], data_grafik["operasional"], label="Operasional", linestyle='-', marker='s')

# Plot data pendapatan lain
plt.plot(data_grafik["tahun"], data_grafik["pendapatan_lain"], label="Pendapatan Lain", linestyle='-.', marker='d')


# Mengatur label, judul, dan legenda
plt.legend(loc="upper left")
plt.xlabel("Tahun")
plt.ylabel("Dalam Jutaan Rp")
plt.title("Data Rekapitulasi Tahunan")
plt.grid(True)  # Menambahkan grid untuk visualisasi yang lebih baik

# Menampilkan grafik
plt.show()
```
